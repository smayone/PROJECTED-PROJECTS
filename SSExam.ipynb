Create a function that reverses a string. For example, given the string **forward**, your function should return **drawrof**. Assume that all inputs are strings.


```
def reverse(s):
  str = ""
  for i in s:
    str = i + str
  return str
s = "for" + "ward"
print(s)
reverse(s)
```
